```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
obs_table = pd.read_table('../getImages/asteroid_families/mba_fam_data.txt')
all_fam_table = pd.read_table('../getImages/asteroid_families/all_data_status_3.txt')
all_table = pd.read_table('all_objects_table.txt')
all_none_table = pd.read_table('../getImages/asteroid_families/all_data_status_0.txt')
```

```python
with sns.axes_style("ticks"):
    plt.scatter(all_table['semimajor_axis'], all_table['eccentricity'], color="#666666", alpha=1, s=1, marker='o', label='all')
    plt.scatter(all_none_table['semimajor_axis'], all_none_table['eccentricity'], color="#1376EF", alpha=1, s=1, marker='o', label='all_none')
    plt.scatter(all_fam_table['semimajor_axis'], all_fam_table['eccentricity'], color="#999999", alpha=0.4, s=1, marker='o', label='all_fam')    
    plt.scatter(obs_table['semimajor_axis'], obs_table['eccentricity'], color="#EF1313", alpha=1, s=1, marker='o', label='obs_fam')
    
    plt.plot([2.5, 2.5], [0, 1], color='k', linestyle='--', linewidth=1)
    plt.plot([2.06, 2.06], [0, 1], color='k', linestyle='--', linewidth=1)
    plt.plot([2.82, 2.82], [0, 1], color='k', linestyle='--', linewidth=1)
    plt.plot([3.27, 3.27], [0, 1], color='k', linestyle='--', linewidth=1)
    plt.plot([1.523, 1.523], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([5.204, 5.204], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.text(2.515, 0.9, '3:1', fontsize=12)
    plt.text(2.075, 0.9, '4:1', fontsize=12)
    plt.text(2.835, 0.9, '5:2', fontsize=12)
    plt.text(3.285, 0.9, '2:1', fontsize=12)
    plt.text(1.538, 0.9, 'a$_{mars}$', fontsize=12)
    plt.text(5.219, 0.9, 'a$_{jup}$', fontsize=12)
    
    plt.ylim(0,1)
    plt.ylabel('eccentricity')
    plt.xlim(1, 6)
    plt.xlabel('semimajor axis (AU)')
    plt.legend()
```

```python
with sns.axes_style("ticks"):
    plt.scatter(all_table['semimajor_axis'], all_table['inclination'], color="#666666", alpha=1, s=1, marker='o', label='all')
    plt.scatter(none_table['semimajor_axis'], none_table['inclination'], color="#1376EF", alpha=1, s=1, marker='o', label='all_none')
    plt.scatter(all_fam_table['semimajor_axis'], all_fam_table['inclination'], color="#999999", alpha=0.4, s=1, marker='o', label='all_fam')        
    plt.scatter(obs_table['semimajor_axis'], obs_table['inclination'], color="#EF1313", alpha=1, s=1, marker='o', label='obs_fam')
    plt.plot([2.5, 2.5], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([2.06, 2.06], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([2.82, 2.82], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([3.27, 3.27], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([1.523, 1.523], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.plot([5.204, 5.204], [0, 35], color='k', linestyle='--', linewidth=1)
    plt.text(2.515, 32, '3:1', fontsize=12)
    plt.text(2.075, 32, '4:1', fontsize=12)
    plt.text(2.835, 32, '5:2', fontsize=12)
    plt.text(3.285, 32, '2:1', fontsize=12)
    plt.text(1.538, 32, 'a$_{mars}$', fontsize=12)
    plt.text(5.219, 32, 'a$_{jup}$', fontsize=12)
    plt.ylim(0, 35)
    plt.ylabel('inclination')
    plt.xlim(1, 6)
    plt.legend()
```

```python
with sns.axes_style("ticks"):
    pg = sns.PairGrid(obs_table, x_vars=["semimajor_axis"], y_vars=["eccentricity", "inclination"], 
                      aspect=4, hue="occurance", palette="coolwarm")
    plt.xlim(2.15, 4)
    pg.axes[0][0].set_ylim([0,0.35])
    plt.ylim(0,35)  
    pg.map(plt.scatter, alpha=0.7) 
    pg.add_legend(title='occurance')
    plt.show()
```

```python
with sns.axes_style("ticks"):
    pg = sns.PairGrid(obs_table, x_vars=["semimajor_axis"], y_vars=['occurance'], aspect=4)
    plt.xlim(2.15, 4)
    plt.ylim(0,35)  
    pg.map(plt.scatter) 
    plt.show()
```

```python
with sns.axes_style("ticks"):
    pg = sns.PairGrid(obs_table, x_vars=["inclination"], y_vars=['occurance'], aspect=4)
    plt.xlim(0, 35)
    plt.ylim(0, 35)  
    pg.map(plt.scatter) 
    plt.show()
```

```python
with sns.axes_style("ticks"):
    pg = sns.PairGrid(obs_table, x_vars=["eccentricity"], y_vars=['occurance'], aspect=4)
    plt.xlim(0, 0.3)
    plt.ylim(0, 35)  
    pg.map(plt.scatter) 
    plt.show()
```

```python
ecc_lt_0_45 = all_fam_table.query('eccentricity < 0.45')

with sns.axes_style('ticks'):
    plt.hist(ecc_lt_0_45['eccentricity'].values, 100, histtype="stepfilled", alpha=0.4, log=True, label='all e<0.45')
    plt.hist(obs_table['eccentricity'].values, 100, histtype="stepfilled", alpha=0.3, log=True, label='observed')
    plt.legend()
```

```python
inc_lt_40 = all_fam_table.query('inclination < 40')

with sns.axes_style('ticks'):
    plt.hist(inc_lt_40['inclination'].values, 50, histtype="stepfilled", alpha=0.4, log=True, label='all i<40')  
    plt.hist(obs_table['inclination'].values, 50, histtype="stepfilled", alpha=0.3, log=True, label='observed')
    plt.legend()
```
