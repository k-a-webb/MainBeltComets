```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from astropy.io import ascii
from astropy.table import Table, Column
```

### Read in comet data and parse file

```python
infile = open("CometEls.txt", "rw+")

numlist = []
qlist = []
ecclist = []
inclist = []
namelist = []


for line in infile.readlines():
    s = line #str(infile.readlines(line))
    #print s

    # number
    number = s[0:12].strip()
    # perihelion distance (AU)
    q = float(s[30:40].strip())
    # orbital eccentricity
    ecc = float(s[41:50].strip())
    # inclination, J@00.0 (degrees)
    inc = float(s[71:80].strip())
    # designation and name
    name = s[102:159].strip()
    numlist.append(number)
    qlist.append(q)
    ecclist.append(ecc)
    inclist.append(inc)
    namelist.append(name)

    
tabledata = Table([numlist, qlist, ecclist, inclist, namelist], 
          names=('number', 'perihelion', 'eccentricity', 'inclination', 'name'))

infile.close()

tabledata["semimajor_axis"] = tabledata['perihelion']/(1 - tabledata['eccentricity'])  # a=q/(1-e)

table = pd.DataFrame(np.array(tabledata))

#print table
```

```python
ecc = table['eccentricity']
q = table['perihelion']
i = table['inclination']
a = table['semimajor_axis']
```

```python
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    plt.scatter(a, ecc)
    plt.ylim(0,1.2)
    plt.ylabel('eccentricity')
    plt.xlim(0, 20)

    plt.subplot(212)
    plt.scatter(a, i)
    plt.ylim(0,14)
    plt.ylabel('inclination')
    plt.xlim(0, 20)
    plt.xlabel('semimajor axis (AU)')
```

##### Learning how to query a table...

```python
num_list
# print 'I am a fish of {} scales and {} fins'.format(23, 6)

mbc_qlist = []
mbc_alist = []
mbc_elist = []
mbc_ilist = []

for num in num_list:
    # query = 'number == "{}"'.format(num)    
    object = table.query('number == "{}"'.format(num))
    q1 = float(object["perihelion"])
    a1 = float(object["semimajor_axis"])
    inc1 = float(object["inclination"])
    ecc1 = float(object["eccentricity"])
    mbc_qlist.append(q1)
    mbc_alist.append(a1)
    mbc_elist.append(ecc1)
    mbc_ilist.append(inc1)

mbcdata = {'name': num_list, 'perihelion': mbc_qlist, \
           'eccentricity': mbc_elist, 'inclination': mbc_ilist, \
           'semimajor_axis': mbc_alist}
mbctable = pd.DataFrame(mbcdata)
mbctable["data_type"] = 'Iden. Comet'
```
