```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from astropy.io import ascii
from astropy.table import Table, Column
```

### Open and parse CometEls data

```python
comet_num = []
comet_q = []
comet_e = []
comet_i = []
comet_name = []

with open("CometEls.txt", "rw+") as infile:
    for line in infile.readlines():
        comet_num.append(line[0:12].strip())         # number
        comet_q.append(float(line[30:40].strip()))   # perihelion distance (AU)
        comet_e.append(float(line[41:50].strip()))   # orbital eccentricity
        comet_i.append(float(line[71:80].strip()))   # inclination, J@00.0 (degrees)
        comet_name.append(line[102:159].strip())     # designation and name

    comet_data = Table([comet_num, comet_q, comet_e, comet_i], 
              names=('number', 'perihelion', 'eccentricity', 'inclination'))

comet_data["semimajor_axis"] = comet_data['perihelion']/(1 - comet_data['eccentricity'])  # a=q/(1-e)
comet_data["data_type"] = 'Comet'

comet_table = pd.DataFrame(np.array(comet_data))
```

### Find objects in CometEls data set with orbital elements of Lixiaohua family

#### Lixiaohua: a = 3.156 AU, e = 0.242, i = 10.97 degrees

```python
in_rangeComets = comet_table.query('(3.1 < semimajor_axis < 3.2) & (0.22 < eccentricity < 0.26) & (10.5 < inclination < 11.5)')

if len(in_rangeComets) == 0:
    print " no Lixiaohua family objects in MPCORB file "
```

### Open and parse MPCORB data

```python
mba_num = []
mba_q = []
mba_e = []
mba_i = []
mba_a = []
mba_name = []

with open('MPCORB.DAT', 'r') as infile:
    # skip header files
    x = 0
    for x in range (0, 41):
        next(infile)
        x += 1
    
    for line in infile:
        if len(line.split()) > 0:
            mba_num.append(line[0:8].strip())
            mba_name.append(line[107:117].strip())
            mba_q.append(float(line[37:47].strip()))
            mba_e.append(float(line[70:80].strip()))
            mba_i.append(float(line[59:69].strip()))
            mba_a.append(float(line[92:104].strip()))
            
mba_data = Table([mba_num, mba_q, mba_e, mba_i, mba_a], 
              names=('number', 'perihelion', 'eccentricity', 'inclination', 'semimajor_axis'))
mba_data['data_type'] = 'Main_belt_asteroid'
mba_table = pd.DataFrame(np.array(mba_data))

#Jupiter range
in_rangeMB = mba_table.query('(1.80 < semimajor_axis < 3.60)')
```

### Find objects in CometEls data set with orbital elements of Lixiaohua family

```python
in_rangeMPCORB = mba_table.query('(3.1 < semimajor_axis < 3.2) & (0.22 < eccentricity < 0.26) & (10.5 < inclination < 11.5)')
# returns 420 objects, arbitrary
in_rangeMPCORB2 = mba_table.query('(3.12 < semimajor_axis < 3.17) & (0.22 < eccentricity < 0.26) & (10.5 < inclination < 11.5)')
# returns 299 objects, overlapping resonances
in_rangeMPCORB3 = mba_table.query('(3.143 < semimajor_axis < 3.150) & (0.22 < eccentricity < 0.26) & (10.5 < inclination < 11.5)')
# returns 50 objects, extended chaotic zone
in_rangeMPCORB4 = mba_table.query('(3.12 < semimajor_axis < 3.17) & (0.22 < eccentricity < 0.26) & (10.8 < inclination < 11.2)')
# returns 135 objects, overlapping resonances

if len(in_rangeMPCORB) == 0:
    print " no Lixiaohua family objects in MPCORB file "
```

### Find confirmed comets (list in mbc.txt) in CometEls data set 

```python
with open("mbc.txt") as infile:
    filestr = infile.read()
aa_num = filestr.split('\n')

aa_q = []
aa_e = []
aa_i = []
aa_a = []

for num in aa_num:
    aa = comet_table.query('number == "{}"'.format(num))
    aa_a.append(float(aa['semimajor_axis']))        
    aa_q.append(float(aa['perihelion']))             
    aa_e.append(float(aa['eccentricity']))   
    aa_i.append(float(aa['inclination']))   

aa_data = Table([aa_num, aa_q, aa_e, aa_i, aa_a], 
              names=('number', 'perihelion', 'eccentricity', 'inclination', 'semimajor_axis'))

aa_data["data_type"] = 'Active_asteroid'

aa_table = pd.DataFrame(np.array(aa_data))
```
